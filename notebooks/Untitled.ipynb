```python
import numpy as np
from numpy import random
from scipy.stats import norm
import matplotlib.pyplot as plt
```

```python
def fibonacci(n):
    u_0 = 0
    u_1 = 1
    for i in range(n-1):
        s = u_0 + u_1
        u_0 = u_1
        u_1 = s
    return s

print(fibonacci(3))
```

```python
from random import choice
import numpy as np
from scipy import interpolate
import matplotlib.pyplot as plt

#Fibonacci strat

POT = 50
cible_axis = []
prob_axis = []
ITERATIONS = 10000

for i in range(10, 101):

    Jetons = POT

    CIBLE = Jetons+i

    cible_axis.append(CIBLE)

    victoires = 0

    for _ in range(ITERATIONS):

        fibo_1 = 0

        fibo_2 = 1

        Jetons = POT

        while Jetons<CIBLE and Jetons-fibo_2>0:

            mise = fibo_2

            Jetons-=mise

            sortie = choice([i for i in range(37)])

            if sortie%3!=0 or sortie==0: #perdu

                tmp = fibo_2

                fibo_2 = fibo_1+fibo_2  

                fibo_1 = tmp

            else: #gagné

                Jetons+=3*mise

                fibo_1 = 0

                fibo_2 = 1

        if Jetons>=POT:

            victoires+=1

    prob_axis.append(victoires*100/ITERATIONS)

 

# x_new, bspline, y_new

x_new = np.linspace(min(cible_axis), max(cible_axis), (max(cible_axis)-min(cible_axis))//3)

x_new = x_new.astype(int)

bspline = interpolate.make_interp_spline(cible_axis, prob_axis)

y_new = bspline(x_new)

#intersect

idx = np.argwhere(np.diff(np.sign(60-y_new))).flatten()

 

# Plot the new data points

plt.plot(x_new, y_new)

plt.plot(x_new[idx], y_new[idx], 'ro')

plt.xticks(x_new, rotation=60)

plt.grid()

plt.show()
```
